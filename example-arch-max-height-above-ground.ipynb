```python
%matplotlib inline

import json
import matplotlib.pyplot as plt
import pdal

pipeline = {
    "pipeline": [
        "arch.las",
        {
            "type": "filters.hag"
        }
    ]
}

p = pdal.Pipeline(json.dumps(pipeline))
p.validate()
count = p.execute()
```

```python
print('The gateway arch is about %.2f m tall'
      % p.arrays[0]['HeightAboveGround'].max())
```

```python
plt.figure(figsize=(20,10));
plt.subplot(121); plt.axis('equal'); plt.title('Raw Elevation');
plt.scatter(p.arrays[0]['Y'], p.arrays[0]['Z'], s=0.1);
plt.subplot(122); plt.axis('equal'); plt.title('Height Above Ground');
plt.scatter(p.arrays[0]['Y'], p.arrays[0]['HeightAboveGround'], s=0.1);
```
